```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("housing.csv")
```

```python
df['income cat'] = pd.cut(df['median_income'], bins = [0,1.5,3.0,4.5,6.0, np.inf], labels = [1,2,3,4,5])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(df, df['income cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[train_index]
```

```python
for sett in (strat_train_set, strat_test_set):
    sett.drop("income cat", axis = 1, inplace = True)
```

```python
df = strat_train_set.copy()
```
